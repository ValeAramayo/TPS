```python
# Códigos y graficos de respuesta de módulo y fase #
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import bode, TransferFunction

def plot_bode(w, mag, phase):
    """Función para graficar el diagrama de Bode."""
    plt.figure()
    
    # Magnitud
    plt.subplot(2, 1, 1)
    plt.semilogx(w, mag)
    plt.xlabel("Frecuencia [rad/s]")
    plt.ylabel("Magnitud [dB]")
    plt.title("Diagrama de Bode - Magnitud")
    plt.grid(True)
    
    # Fase
    plt.subplot(2, 1, 2)
    plt.semilogx(w, phase)
    plt.xlabel("Frecuencia [rad/s]")
    plt.ylabel("Fase [°]")
    plt.title("Diagrama de Bode - Fase")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Definir parámetros del sistema
w0 = 1  # Frecuencia natural
q = 1   # Factor de calidad

# Defino la función de transferencia H(s) para le punto a
my_tf = TransferFunction([w0**2, 0], [1, w0/q, w0**2])
# Defino la función de transferencia H(s) para le punto b
my_tf2 = TransferFunction([1,0,0], [1, w0/q, 1/w0**2])


# Obtengo la respuesta en frecuencia
w, mag, phase = bode(my_tf)
w2, mag2, phase2 = bode(my_tf2)
# Llamar a la función para graficar
plot_bode(w, mag, phase)
plot_bode(w2, mag2, phase2)

```

Por otro lado, el ancho de banda de un filtro pasa-banda está definido por el intervalo entre su frecuencia de corte inferior y su frecuencia de corte superior, es decir, los puntos donde la atenuación es de -3 dB. # Bonus #
Hay dos formas para conectar los filtros en cascada. La primera forma es que la salida del filtro pasa altos sea la entrada del filtro pasa banda y la otra forma es que la salida del pasa banda sea la señal de entrada del filtro pasa altos. 

- Si se pone primero el pasa-altos, solo pasan las frecuencias altas antes de que el pasa banda haga su trabajo. Eso hace que el resultado siga siendo un pasa banda pero con la frecuencia de corte inferior ajustada por el primer filtro. O sea, se filtran más las bajas frecuencias y la banda que pasa queda más controlada. Por otro lado, si se conectan al revés (pasa banda primero y después el pasa altos), se puede terminar matando la señal porque la frecuencia de de corte del pasa altos es mayor que la banda del pasa banda. En ese caso, casi no pasaría nada de señal. Concluimos entonces que conviene primero el pasa-altos y después el pasa-banda, así se mantiene una banda pasante bien definida sin riesgo de perder todo.
